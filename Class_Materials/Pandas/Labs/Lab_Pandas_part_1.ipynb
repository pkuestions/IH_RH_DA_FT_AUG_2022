### Cleaning and wrangling data

- Some car brand names refer to the same brand. Replace all brand names that contain the word "Dutton" for simply "Dutton". If you find similar examples, clean their names too. Use `loc` with boolean indexing.

- Convert CO2 Emissions from Grams/Mile to Grams/Km

- Create a binary column that solely indicates if the transmission of a car is automatic or manual. Use `pandas.Series.str.startswith` and .

- convert MPG columns to km_per_liter

```python
df_vehicles = pd.read_csv("data/vehicles.csv")
```

```python
def ReplaceColValue(col, find, replace):
    df_vehicles.loc[df_vehicles[df_vehicles[col].str.contains(find)].index,col] = replace
    
    if col not in df_vehicles.columns:
        raise ValueError("Column not in dataframe")
        
    return df_vehicles
```

```python
ReplaceColValue("Make", "smart", "Smart")
```

```python
## Your Code here
df_vehicles.loc[df_vehicles[df_vehicles["Make"].str.contains("Dutton")].index,"Make"] = "Dutton"
```

```python
Gram_per_Km = list(map(lambda x: x * (1/1.60934), df_vehicles["CO2 Emission Grams/Mile"]))
df_vehicles.insert(len(df_vehicles.columns)-1, "CO2 Emission Grams/Km", Gram_per_Km)
df_vehicles
```

```python
df_vehicles.insert(list(df_vehicles.columns).index("Transmission")+1, 
                   "Transmission_Auto", df_vehicles["Transmission"].str.startswith("Auto"))
df_vehicles
```

```python
df_vehicles["City MPG"] = list(map(lambda x: x * (1.60934/3.78541), df_vehicles["City MPG"]))
df_vehicles["Highway MPG"] = list(map(lambda x: x * (1.60934/3.78541), df_vehicles["Highway MPG"]))
df_vehicles["Combined MPG"] = list(map(lambda x: x * (1.60934/3.78541), df_vehicles["Combined MPG"]))

df_vehicles.rename(columns={"City MPG": "City km/L", "Highway MPG": "Highway km/L", "Combined MPG": "Combined km/L"},
                   inplace=True)
df_vehicles
```

Converting Grams/Mile to Grams/Km

1 Mile = 1.60934 Km

Grams/Mile * Mile/Km -> Grams/Mile * 1 Mile/1.60934Km

$$ \frac{Grams}{Mile} * \frac{Mile}{Km} $$

$$ \frac{Grams}{Mile} * \frac{1 Mile}{1.60934Km}  $$

convert MPG columns to km_per_liter

MPG = Miles/Gallon -> Km/Liter

1 Mile = 1.60934 Km

1 Gallon = 3.78541 Liters

$$ \frac{Miles}{Gallon} -> \frac{Miles}{Gallon} * \frac{Km}{Miles} * \frac{Gallon}{Liters}$$

$$ \frac{Miles}{Gallon} -> \frac{Miles}{Gallon} * \frac{1.60934Km}{ 1Miles} * \frac{1 Gallon}{3.78541 Liters}$$

* ( 1.60934 / 3.78541 )

### Gathering insights:

- How many car makers are there? How many models? Which car maker has the most cars in the dataset?

- When were these cars made? How big is the engine of these cars?

- What's the frequency of different transmissions, drivetrains and fuel types?

- What's the car that consumes the least/most fuel?

```python
# Your Code here
from collections import Counter

Amount_of_Carmakers = len(set(df_vehicles["Make"]))
Amount_of_Models = len(set(df_vehicles["Model"]))
Most_Cars = df_vehicles["Make"].value_counts()[0]
Most_Cars_Carmaker = Counter(df_vehicles["Make"]).most_common()[0][0] # df_vehicles["Make"].mode()[0]
#Most_Cars = Counter(df_vehicles["Make"]).most_common()[0][1]
Most_Cars_Carmaker_Period = [df_vehicles[df_vehicles["Make"]==Most_Cars_Carmaker]["Year"].min(),
                            df_vehicles[df_vehicles["Make"]==Most_Cars_Carmaker]["Year"].max()]
Most_Cars_Carmaker_Engine = [df_vehicles[df_vehicles["Make"]==Most_Cars_Carmaker]["Cylinders"].describe()["min"],
                            df_vehicles[df_vehicles["Make"]==Most_Cars_Carmaker]["Cylinders"].describe()["max"],
                            df_vehicles[df_vehicles["Make"]==Most_Cars_Carmaker]["Cylinders"].describe()["50%"]]

print(f"Number of car makers = {Amount_of_Carmakers}.")
print(f"Number of car models = {Amount_of_Models}.")
print(f"Car maker with most cars in dataset is {Most_Cars_Carmaker} with {Most_Cars} cars.")
print(f"These cars were made between {Most_Cars_Carmaker_Period[0]}-{Most_Cars_Carmaker_Period[1]}.")
print(f"These cars have between {int(Most_Cars_Carmaker_Engine[0])}-{int(Most_Cars_Carmaker_Engine[1])} cylinders (median = {int(Most_Cars_Carmaker_Engine[2])}).")

Transmission_Frequency = df_vehicles["Transmission"].value_counts()
Drivetrain_Frequency = df_vehicles["Drivetrain"].value_counts()
Fuel_Type_Frequency = df_vehicles["Fuel Type"].value_counts()

min_consum = df_vehicles["Fuel Barrels/Year"].describe()["min"]
max_consum = df_vehicles["Fuel Barrels/Year"].describe()["max"]
car_min_consum = df_vehicles[df_vehicles["Fuel Barrels/Year"]==min_consum].iloc[0,:2]["Make"] + " " +\
df_vehicles[df_vehicles["Fuel Barrels/Year"]==min_consum].iloc[0,:2]["Model"]
car_max_consum = df_vehicles[df_vehicles["Fuel Barrels/Year"]==max_consum].iloc[0,:2]["Make"] + " " +\
df_vehicles[df_vehicles["Fuel Barrels/Year"]==max_consum].iloc[0,:2]["Model"]

print(f"Car that consumes the least fuel is {car_min_consum}.")
print(f"Car that consumes the most fuel is {car_max_consum}.")
```

Do cars with automatic transmission consume more fuel than cars with manual transmission on average?

```python
## Your Code is here
Consum_Automatic = round(df_vehicles[df_vehicles["Transmission_Auto"]==True]["Fuel Barrels/Year"].mean(),2)
Consum_Manual = round(df_vehicles[df_vehicles["Transmission_Auto"]==False]["Fuel Barrels/Year"].mean(),2)
Consum_diff = int((Consum_Manual-Consum_Automatic)/Consum_Manual*100)

print(f"Automatic cars drive on average {Consum_Automatic} km/L, whereas manual cars drive on average {Consum_Manual} km/L.")
print(f"Hence, automatic cars consume on average {Consum_diff} % more fuel.")
```
